```python
import yfinance as yf
import pandas as pd
import math

ticker_list = 'tickers.txt'
tickers = []

# Read the text file containing the tickers
with open(ticker_list) as f:
  for line in f.readlines():
    tickers.append(line.strip())

# Create an empty dataframe to store the results
results = pd.DataFrame(columns=['Ticker', 'Correlation'])

# Define the date range for which to retrieve data
start = '2018-01-11'
end = '2023-01-11'

# Retrieve historical data for the S&P 500 index
sp500_data = yf.download('^GSPC', start=start, end=end)

# Calculate the S&P 500's daily returns
sp500_returns = sp500_data['Adj Close'].pct_change()

# Iterate through the list of tickers
for ticker in tickers:
    # Retrieve historical stock data for the stock
    stock_data = yf.download(ticker, start=start, end=end)

    # Calculate the stock's daily returns
    stock_returns = stock_data['Adj Close'].pct_change()

    # Compare the stock's returns to the S&P 500's returns
    correlation = stock_returns.corr(sp500_returns)

    # Append the ticker and comparison to the dataframe only if correlation is not NaN
    if not math.isnan(correlation):
      results = results.append({'Ticker': ticker, 'Correlation': correlation}, ignore_index=True)

# Display the dataframe
print(results)
```
